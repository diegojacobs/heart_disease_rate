Normalize data using Min-Max normalization.
Columns:
- demo__birth_rate_per_1k
- demo__death_rate_per_1k
- health__air_pollution_particulate_matter
- health__homicides_per_100k
- health__motor_vehicle_crash_deaths_per_100k
- health__pop_per_dentist
- health__pop_per_primary_care_physician

```python
train_data = train_data.assign(demo__birth_rate_per_1k=lambda x: (x.demo__birth_rate_per_1k - train_data['demo__birth_rate_per_1k'].min())/(train_data['demo__birth_rate_per_1k'].max() - train_data['demo__birth_rate_per_1k'].min()))
train_data = train_data.assign(demo__death_rate_per_1k=lambda x: (x.demo__death_rate_per_1k - train_data['demo__death_rate_per_1k'].min())/(train_data['demo__death_rate_per_1k'].max() - train_data['demo__death_rate_per_1k'].min()))
train_data = train_data.assign(health__air_pollution_particulate_matter=lambda x: (x.health__air_pollution_particulate_matter - train_data['health__air_pollution_particulate_matter'].min())/(train_data['health__air_pollution_particulate_matter'].max() - train_data['health__air_pollution_particulate_matter'].min()))
train_data = train_data.assign(health__homicides_per_100k=lambda x: (x.health__homicides_per_100k - train_data['health__homicides_per_100k'].min())/(train_data['health__homicides_per_100k'].max() - train_data['health__homicides_per_100k'].min()))
train_data = train_data.assign(health__motor_vehicle_crash_deaths_per_100k=lambda x: (x.health__motor_vehicle_crash_deaths_per_100k - train_data['health__motor_vehicle_crash_deaths_per_100k'].min())/(train_data['health__motor_vehicle_crash_deaths_per_100k'].max() - train_data['health__motor_vehicle_crash_deaths_per_100k'].min()))
train_data = train_data.assign(health__pop_per_dentist=lambda x: (x.health__pop_per_dentist - train_data['health__pop_per_dentist'].min())/(train_data['health__pop_per_dentist'].max() - train_data['health__pop_per_dentist'].min()))
train_data = train_data.assign(health__pop_per_primary_care_physician=lambda x: (x.health__pop_per_primary_care_physician - train_data['health__pop_per_primary_care_physician'].min())/(train_data['health__pop_per_primary_care_physician'].max() - train_data['health__pop_per_primary_care_physician'].min()))
```

244 - 377 = 1 -> Middle
378 - 512 = 2 -> High

```python
for i in range(0,len(train_data)):
    if(train_data.iloc[i,34] < 244):
        train_data.iloc[i,34] = 0
    elif(train_data.iloc[i,34] > 243 and train_data.iloc[i,34] < 377):
        train_data.iloc[i,34] = 1
    else:
        train_data.iloc[i,34] = 2
```
